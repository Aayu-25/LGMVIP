<b>STEP 2 :- Loading the image</b><br> The image that is to be processed is loaded using imread() method. 

```python
image = cv2.imread("C:\\Users\\Hp\\Downloads\\flower.jpg",1)
cv2.imshow("ROSE",image)
#filename = 'Flower.jpg'
```

```python
cv2.imwrite('flower.jpg',image)
```

```python
Image(filename="flower.jpg")
```

<b>STEP 3 :- Converting an image into gray_scale image</b><br> The input image is converted into equivalent grey-scale using cv2.cvtColor. Converting an image into grayscale gives a black & white pixels in the image which is used for creating a pencil sketch. 

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

cv2.imwrite('gray_image.jpg',gray_image)
Image(filename="gray_image.jpg")
```

<b>STEP 4 :- Inverting the image</b><br>The grayscale image is converted to an inverted version of image i.e. lighter regions are converted to dark and vice versa so that the edges to create a pencil sketch are found.

```python
inverted_image = 255 - gray_image
cv2.imwrite('inverted_image.jpg',inverted_image)
Image(filename="inverted_image.jpg")
```

<b>STEP 5 :- Smoothing the image </b><br>Next step is to blur the image by using the GaussianBlur Function in OpenCV. We have used the gaussian blur technique with 21 x 21 pixel and the default sigma values filter on the image to smoothen our image. By increasing the filter size, we can create thin lines for our sketch and it is used to reduce the noise in the image. 

```python
blurred = cv2.GaussianBlur(inverted_image, (21, 21), sigmaX=0, sigmaY=0)
```

<b>STEP 6 :- Obtaining the final sketch</b><br>Next step is the inversion of blurred image.The cv2.divide() function is used for dividing the greyscale value of the image by the inverse of blurred image value which highlights the boldest edges.

```python
inverted_blurred = 255 - blurred
pencil_sketch = cv2.divide(gray_image, inverted_blurred, scale=256.0)
cv2.imwrite('pencil_sketch.jpg',pencil_sketch)
Image(filename="pencil_sketch.jpg")
```

<b>STEP 7 :- FINAL SKETCH</b><br>Using the imshow() function, the final sketch can be printed.

```python
cv2.imshow("original image", image)
cv2.imwrite('flower.jpg',image)
Image(filename="flower.jpg")
cv2.imshow("pencil sketch", pencil_sketch)
cv2.imwrite('pencil_sketch.jpg',pencil_sketch)
Image(filename="pencil_sketch.jpg")
#cv2.waitKey(0)
```
